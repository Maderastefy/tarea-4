```python
# Pregunta 1. Metodo grafico
import numpy as np
import matplotlib.pyplot as plt
from math import exp

def h(c):
  return (668.06/c)*(1-np.exp(-0.14684*c))-40

c = np.linspace(0.1,50,100)

funtion = h(c)

# grafica
plt.plot(c, funtion)
plt.axhline(y=0, color='r', linestyle='--')
plt.axvline(x=15, color='g', linestyle='--')
plt.title("metodo grafica")
plt.grid()
plt.show()

# Teniendo en cuenta que V=mg/c *(1-e**(-ct/m)).
# Entonces el coeficiente de resistencia, c. debe ser aproximadamente 15 como se ve en la grafica para que la velocidad terminal sea cercano a 40 m/s
```

```python
# Pregunta 5

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
from math import exp

#Metodo analitico

def f(y, t):
   return y*((t**2)-1.1)

#definimos las condiciones
fo = 1, 0
t = np.linspace(0, 2, 100)

sol = integrate.odeint(f, fo, t)
print(sol)

#valor real

def r(t):
   return np.exp((t**3)/3-(11/10)*t)
```

```python
#Metodo de EULER

import numpy as np
import matplotlib.pyplot as plt

def funcion(t, y, k):
  return y*((t**2)-k)

#configurar parametros
k = np.float(1.1)
h=0.5
t = np.arange(0, 2, h)

#inicia arrays para almacenar
y = np.zeros_like(t)
y[0] = 1 #condicion inicial

#metodo euler

for i in range(1,len(t)):
   y[i]=y[i-1] + h * funcion(t[i], y[i], k)
   t[i]=t[i]+h

print(t[i], y[i])
plt.scatter(t,y)
```
